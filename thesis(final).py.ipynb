```python
X=pd.DataFrame(columns=['Age','Experience','Income','Family','CCAvg','Education','Mortgage','Securities Account','CD Account','CreditCard','Online'],data=df)
y=df['Personal Loan']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.3)
X_train.head(5)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree= DecisionTreeClassifier(max_leaf_nodes=5)
dtree.fit(X_train,y_train)
predictions= dtree.predict(X_test)
print('The Score on the training dataset with decision tree classifier is : ',dtree.score(X_train,y_train))
print('The Score on the testing dataset with decision tree classifier is : ',dtree.score(X_test,y_test))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import metrics
a = metrics.accuracy_score(y_test, predictions)*100
print("Decision Tree Classifier model accuracy(in %):", a)
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn import tree
fn=['Age','Experience','Income','Family','CCAvg','Education','Mortgage','Securities Account','CD Account','CreditCard','Online']
cn=['Deny Offer','Accpect Offer']
plt.figure(figsize=(25,10), facecolor = 'gray')
dtree_plot = tree.plot_tree(dtree,feature_names=fn,class_names=cn, filled =True, rounded = True, fontsize = 16)
plt.savefig('detree',dpi=400)
plt.show(dtree_plot)
```

```python
from sklearn.naive_bayes import GaussianNB 
gnb = GaussianNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test) 
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import metrics
b =  metrics.accuracy_score(y_test, y_pred)*100
print("Gaussian Naive Bayes model accuracy(in %):",b)
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn import svm
svm_clf = svm.SVC(kernel = 'rbf', random_state =4,probability =True)
svm_clf.fit(X_train,y_train)
predict = svm_clf.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import metrics
c = metrics.accuracy_score(y_test, predict)*100
print("Support Vector Machine Classifier model accuracy(in %):", c)
print(confusion_matrix(y_test,predictions))
```

```python
d = np.random.normal(a,1,500)
e = np.random.normal(b,1,500)
f = np.random.normal(c,1,500)
results = [d,e,f]
names = ['Decision Tree','Gaussian Naive Bayes','Support Vector Machine']
fig = plt.figure(figsize=(10,7))
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
bp = ax.boxplot(results)
ax.set_xticklabels(names)
plt.show(bp)
```

```python
y_pred = gnb.fit(X_train, y_train)
nb_probs =y_pred.predict_proba(X_test)
svm_probs = svm_clf.predict_proba(X_test)
dt_probs = dtree.predict_proba(X_test)

nb_probs = nb_probs[:,-1]
dt_probs = dt_probs[:,-1]
svm_probs = svm_probs[:,-1]

from sklearn.metrics import roc_curve,roc_auc_score
nb_auc =roc_auc_score(y_test,nb_probs)
dt_auc =roc_auc_score(y_test,dt_probs)
svm_auc =roc_auc_score(y_test,svm_probs)

print("Gaussian Naive Bayes:AUROC =%0.3f" %(nb_auc))
print("Decision Tree:AUROC =%0.3f" %(dt_auc))
print("Support Vector:AUROC =%0.3f" %(svm_auc))

nb_fpr,nb_tpr,_ = roc_curve(y_test,nb_probs)
dt_fpr,dt_tpr,_ = roc_curve(y_test,dt_probs)
svm_fpr,svm_tpr,_ = roc_curve(y_test,svm_probs)

import matplotlib.pyplot as plt
plt.figure(figsize =(15,10))
plt.plot(nb_fpr,nb_tpr,linestyle ='--',label ='Gaussian Naive Bayes (AUROC =%0.3f)'%nb_auc)
plt.plot(dt_fpr,dt_tpr,linestyle ='--',label ='Decision Tree (AUROC =%0.3f)'%dt_auc)
plt.plot(svm_fpr,svm_tpr,linestyle ='--',label ='Support Vector (AUROC =%0.3f)'%svm_auc)

plt.title('ROC Plot')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```
